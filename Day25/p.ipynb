```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as scs
import math
```

```python

x=np.linspace(0,2*np.pi,400)
plt.figure(figsize=(14,4))
plt.plot(x,np.sin(x),label="y=sin(x)")
plt.plot(x,np.sin(2*x),label="y=sin(2x)")
plt.title("(i) y=sin(x) and y=sin(2x)")
plt.grid(True)
plt.show()
```

```python
x=np.linspace(0,2*np.pi,400)
plt.figure(figsize=(14,4))
plt.plot(x,np.sin(x+np.pi/4),label="y=sin(x+pi/4)")

plt.title("(ii) sin(x+pi/4)")
plt.grid(True)
plt.show()
```

```python
x=np.linspace(0,2*np.pi,400)
plt.figure(figsize=(14,4))
plt.plot(x,np.exp(-0.1*x)*np.sin(3*x),label="y=exp(-0.1x)*sin(3x)")

plt.title("(iii) y=exp(-0.1x)*sin(3x)")
plt.grid(True)
plt.show()
```

```python
x=np.linspace(0,2*np.pi,400)
plt.figure(figsize=(14,4))
plt.plot(x,np.sin(x)*np.cos(2*x),label="y=sin(x).cox(2x)")

plt.title("(iii) y=sin(x).cox(2x)")
plt.grid(True)
plt.show()
```
